```python
playerid23 = df23['PlayerId'].tolist()
playerid24 = df24['PlayerId'].tolist()
playerid24[178]
```

```python
test = 'FB98'
if test in playerid24 and playerid23:
    print('true')
else :
    print('false')
```

```python
numberDict = {}
matchesDict = {}
for i in range(200):
    numberDict[i] = df24['Number'][i]
    matchesDict[i] = df24['Matches'][i]
print(numberDict)
print(matchesDict)
```

```python
def acesPerMatch(player):
    if matchesDict[player] == 0:
        return 0
    else:
        numberDict[player] = float(numberDict[player])
        matchesDict[player] = float(matchesDict[player])
        APM = round(numberDict[player] / matchesDict[player], 2)
        return APM
```

```python
if test in playerid24:
    apm24 = acesPerMatch(playerid24.index(test))
    apm23 = acesPerMatch(playerid23.index(test))

perApm = (apm24 / apm23)
print(round(perApm, 2))
print('percent improvement from last year')
```
